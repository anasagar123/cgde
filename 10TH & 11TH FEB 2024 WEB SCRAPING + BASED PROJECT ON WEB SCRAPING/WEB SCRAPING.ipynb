```python
import requests
from bs4 import BeautifulSoup

# Beautiful soup deals with html,xml data.
# In html and xml, so many thing is there like tag,attribute,header,
#   string etc. Through Beautiful soup we can easily search , modify , navigate a 
#   parse tree html,xml files. 
```

```python
url = 'https://www.worldometers.info/world-population/india-population/'
r= requests.get(url)
soup = BeautifulSoup(r.text,'lxml')
print(soup)
```

```python
# if i want to div tags only then
print(soup.div)
# The <div> tag defines a division or a section in an HTML document.
```

```python
print(soup.div.a) # inside the div a tag is there
```

```python
print(soup.div.ul) # inside the div ul(unordered list) tag is there
```

```python
print(soup.find('div'))
```
