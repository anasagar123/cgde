```python
from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder,SystemMessagePromptTemplate,HumanMessagePromptTemplate
from langchain_core.messages import SystemMessage
```

```python
chat_prompt_template = ChatPromptTemplate.from_messages([
    SystemMessage(content="You are a chatbot having a conversation with a human."), # system message i am giving to a bot
    
    # creating a chat_history placeholder(whatever conversation happened we are storing in chat_history)
    MessagesPlaceholder(variable_name="chat_history"),
    # Human Prompt
    HumanMessagePromptTemplate.from_template("{human_input}"),])
```

**Step 3 - Initialize the Memory**

```python
from langchain.memory import ConversationBufferMemory
memory = ConversationBufferMemory(memory_key="chat_history",return_messages=True)
```

**Step 4 - Create a Output Parser**

```python
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
```

**Step 5 - Build a Chain (Another way)**

```python
#    ## we can use this approach also
# chain = chat_template | chat_model | output_parser

# input = {'topic_name':"Logistic Regression"}

# response = chain.invoke(input)

# print(response)
```

```python
                ## Another Approach
from langchain_core.runnables import RunnablePassthrough, RunnableLambda

# define a function to load the message from memory
# here we are storing chat_history also 
def get_messages_from_memory(human_input):
    return memory.load_memory_variables(human_input)['chat_history']


# Define a chain
chain = RunnablePassthrough.assign(chat_history = RunnableLambda(get_messages_from_memory)) | chat_prompt_template | chat_model | output_parser 
# here i am passing history first after that template then model after that output_parser.  
# .assign - means we are connecting with the chat_history
```

**Step 6 - Invoke the chain with human_input and chat_history**

```python
query = {"human_input":"Hi How are you?"}
response = chain.invoke(query)
response
```

**Step 7 - Saving to memory**

```python
# here response is also saving in memory.
# whatever the conversation we are doing it, we save that part as well as
```

```python
memory.save_context(query,{'output':response})
memory.buffer
```

**Step 8 - Run Step 6 and 7 in a loop**

```python

while True:
    query = {"human_input":input("Enter your input here: ")} # here we will pass input.
    print(f"*User:{query['human_input']}")  # user question ask
    if query['human_input'] in ['bye','quit','exit']:
        break
    response = chain.invoke(query)
    print(f"*AI:{response}")  # AI give the response
    memory.save_context(query,{'output':response})  # history save in memory
```

```python
from langchain_community.chat_message_histories import SQLChatMessageHistory # here i am SQLChatMessageHistory

def get_session_message_history_from_db(session_id): # each session generated id
    chat_message_history = SQLChatMessageHistory(session_id,session_id,connection='sqlite:///chat_model/sqlite.db') #connection='sqlite:///chat_model/sqlite.db' here connecting with the sqllite 
    return chat_message_history  
```

```python
# create a chat template

from langchain_core.prompts import ChatPromptTemplate,MessagesPlaceholder
chat_template = ChatPromptTemplate.from_messages([('system',"You are a helpful AI assistant."),
                                                  MessagesPlaceholder(variable_name = "history"),
                                                  ("human","{human_input}")])
                                                   
```

```python
 # Build Chain
chain = chat_template | chat_model
```

```python
from langchain_core.runnables.history import RunnableWithMessageHistory

# Define a chain
conversation_chain = RunnableWithMessageHistory(chain,get_session_message_history_from_db,
                                                input_messages_key="human_input",
                                                history_messages_key = "history")
```

```python
# This is where we configure the session id
user_id = "Anand"
config = {"configurable":{"session_id":user_id}}
input_prompt = {"human_input": "My name is Anand. Can you tell me the Capital of Gujarat?"}
response = conversation_chain.invoke(input_prompt, config=config)
response.content

# in my system sqllite is not there thats why error is coming
```

```python
user_id = "Urmil"
config = {"configurable":{"session_id":user_id}}
input_prompt = {"human_input": "My name is sagar. What are the biggest 3 states in India?"}
response = conversation_chain.invoke(input_prompt, config=config)
print(response.content)
```

```python
def chat_bot(session_id, prompt):
    config={"configurable":{"session_id":user_id}}
    input_prompt = {"human_input": prompt}
    response = conversation_chain.invoke(input_prompt, config=config)
    return response.content    
```

```python
user_id = "Urmil"
input_prompt = "Do you remember my name?"
chat_bot(session_id=user_id, prompt=input_prompt)
# it will give the answer because already conversation happend.
```
