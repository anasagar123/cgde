**Multiple Color at one go**

```python
color_list = [method for method in dir(cv2) if method.startswith('COLOR_') is True]
color_list
```

```python
len(color_list)
```

```python
img = cv2.imread('yash.jpg')
for color in range(len(color_list)):
    col_name = color_list[color]
    colr_code = "cv2."+col_name
    resized = cv2.resize(img,(400,400))
    try:
        rgb_image = cv2.cvtColor(resized,eval(colr_code))
        cv2.imshow(col_name+"_"+str(color),rgb_image)
        cv2.waitkey()
    except Exception as e:
        print("Error on",str(colr_code))
    cv2.destroyAllWindows()

# 374 color is there, These part is not reading thats why some error is coming.
```
