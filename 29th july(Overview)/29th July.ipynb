```python
# Diff between Deep Learning vs Machine Learning :-

#     (1) Data Dependency - Less data use ML(performance would be high if 
#       we will work with less data) and huge data use DL(performance would 
#       be high if data would be huge)

#     (2) Hardware Dependency - In ML we use CPU(central processing unit) but 
#       in DL use GPU(graphical processing unit) or TPU(tensor processing unit),
#       GPU more then 100 times faster then CPU
    
#     (3) S/w Dependency - In Ml S/W use(numpy,pandas,matplotlib,plotly,seaborn,
#          sklearn,scikit and so on),In Deep Learning S/W use(Tensorflow,Keras,Pytorch,NLTK,spacy and so on)

#     (4) Feature Selection - In Ml select Feature Selection manually but in deep learning it will happen automatically

#     (5) Interpretability - In Ml will do manually (missing value,feature selection etc) in ML, But IN DL we will not do  manually anything.

#     (6) Training Time -In ML Training time is very less and test time is more,
#       But In DL Training time is more and test time is less

#     (7) manual vs Automation - manual means Ml and Automation means DL
```
