# Regression Problem

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
df = pd.read_csv('/content/boston.csv')
df.head()
```

```python
df.info()
```

```python
# split the data into x and y
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Mandatory Steps in Deep Learning - Feature Scaling (Normalize the data before building model)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
pd.DataFrame(x_scaler)
```

```python
# split the data into training and test for building model and prediction
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaler, y, test_size=0.25, random_state=101)
```

# Deep Neural Network - MLP Method

```python
import tensorflow as tf
from tensorflow import keras
tf.__version__
```

```python
dnn = keras.models.Sequential()
dnn.add(keras.layers.Dense(units=32, activation='relu'))
dnn.add(keras.layers.Dense(units=32, activation='relu'))
dnn.add(keras.layers.Dense(units=1, activation='linear'))
dnn.compile(optimizer='adam', loss='mean_squared_error')
dnn.fit(x_train, y_train, batch_size=32, epochs=200, validation_data=(x_test, y_test))
```
