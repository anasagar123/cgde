```python
import numpy as np
import cv2
```

```python
def viewImage(image,name_of_window):
    cv2.namedWindow(name_of_window,cv2.WINDOW_NORMAL)
    cv2.imshow(name_of_window,image)
    cv2.waitKey()
    cv2.destroyAllWindows()
```

```python
img = cv2.imread('Modiji.jpg',0) # 0 means balck & white
viewImage(img,"Modi ji")
```

```python
img = cv2.imread('Modiji.jpg',1) # 1 means color(RGB)
viewImage(img,"Modi ji")
```

```python
# increase or decrease the image 30%,20%,60% etc 

scale_percent = 30
width = int(img.shape[1] * scale_percent/100) # shape[1] means second index
height  = int(img.shape[0] * scale_percent/100) # shape[0] means first index
dim = (width,height)

resized_img = cv2.resize(img, dim, interpolation=cv2.INTER_AREA) # interpolation means they will change the image area part width and height 
viewImage(resized_img,"After resizing the image with 30%")
```

**Image rotating**

```python
img.shape
```

```python
# height = 927
# width = 1024
# dimension = 3
```

```python
(h, w, d) = img.shape
center = (w//2, h//2)
M = cv2.getRotationMatrix2D(center, 180, 1.0)
rotated = cv2.warpAffine(img, M, (w,h))
viewImage(rotated, "image after rotation")
```

**Object Detection**

```python
face_classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml") # CascadeClassifier means enabling the xml file
image = cv2.imread('Modiji.jpg')
faces = face_classifier.detectMultiScale(image, 1.3 , 5) # scaleFactor = 1.3 , minSize = 5
# detectMultiScale means in all the angle or edges part they will see or check
if faces is ():
    print("No faces found")
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y),(x+w, y+h),(0,0,255),2) # point1 and point2 = (x+w, y+h) ,color = (0,0,255) means red, thickness = 2
    cv2.imshow("Face Detection",image)
    cv2.waitKey(0)
cv2.destroyAllWindows()
```

**face and eye detection together**

```python
face_classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
eye_classifier = cv2.CascadeClassifier("haarcascade_eye.xml")
image = cv2.imread('trump_henry.jpg')
faces = face_classifier.detectMultiScale(image,1.3,5)
if faces is ():
    print("No faces found")

for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,255),2)
    cv2.imshow('face',image)
    cv2.waitKey(0)
    eye_image = image[y:y+h, x:x+w]
    eyes = eye_classifier.detectMultiScale(eye_image)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(eye_image,(ex,ey),(ex+ew, ey+eh),(255,0,0),2)
        cv2.imshow('eye',image)
        cv2.waitKey(0)
cv2.destroyAllWindows()
    
    
    
    
```
