**Gemini ChatBot**

```python
import google.generativeai as genai
```

```python
# f = open('Gemini_API_key.txt')
# key = f.read()
# genai.configure(api_key=key)
        ## Setting the API Key
f = open('Gemini_API_key.txt')
key = f.read()
genai.configure(api_key = key)
```

```python
for model in genai.list_models():
    print(model.name)
```

```python
model = genai.GenerativeModel(model_name = "models/gemini-1.5-flash")
prompt = "What is Linear Regression"
response = model.generate_content(prompt)
print(response.text)
```

```python
  # chat session
chat = model.start_chat(history=[])
chat
```

```python
# system_instruction - human will ask system will give output

model = genai.GenerativeModel(
        model_name='models/gemini-1.5-flash',
    system_instruction = """ You are a helpful AI assistant.
    You resolve the doubts of student from Data Science domain.
    Your name is "chitti the Robot".In case if someone ask your name tell them.""")
chat = model.start_chat(history=[])
chat
```

```python
while True:
    prompt = input("Enter the Prompt :")
    if prompt.lower() in ["exit","bye"]:
        break
    response = chat.send_message(prompt)
    print("--> ", response.text)
        
```
