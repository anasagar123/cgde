**Loading Llama 2 7B**

```python
# Transformers are a type of deep learning architecture that process input sequences to create output
#   sequences.They learn the context and relationships between the components of a sequence
# AutoTokenizer is also HauggingFace Library,it will allow(Bert model,gpt model,distilbert model etc) all the model to use.no specific package install for specific model.
from transformers import AutoTokenizer #
import transformers
import torch  # pytorch developed by the meta thats why we are using torch model not tensorflow
model = "meta-llama/Llama-2-7b-chat-hf"
```

```python
# load the pretrained (meta-llama/Llama-2-7b-chat-hf) tokenizer

tokenizer = AutoTokenizer.from_pretrained(model,use_auth_token=True) # use_authorization_token is nothing its a api key only
```

```python
# now ready to use the meta model
```

```python
 # make it ready for text generation
pipeline = transformers.pipeline(
     "text-generation",
     model=model,
     torch_dtype=torch.float16,
     device_map="auto"
 )

 # torch_dtype=torch.float16 - which type of model(using pytorch model)
 # device_map="auto" - which type of processor, here it will automatically will take processor(cpu or gpu)
```

```python
# This all part stored in a pipeline.
```

**Define the function that accept the prompt and return the response**

```python
def get_response(prompt):

  sequences = pipeline(prompt,
      do_sample=True,
      return_full_text=False,
      top_k=10,
      #top_p = 0.9,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id,
      max_length=1000,
  )

  return sequences[0]['generated_text']


# ---> do_sample = True  - randomly data i have taken it
# ---> return_full_text = False  - i dont want all the response, i just want to the point only.
# ---> top_k = 10     - give first top 10 probability after that randomly give.
#  example top_k :
#  The cat is -------(sleeping-40%,playing-30%,eating-15%,running-5%,walking-2%,and so on...)
#  top_k = 5
# ---> num_return_sequences = 1   - i want one response only,from sleeping,playing,eating,running...
```

```python
#  The cat is -------(sleeping-40%,playing-30%,eating-15%,running-5%,walking-2%,and so on...)
#  top_p = 0.9 - 0.9 means 90%, top_p add 90%(sleeping-40%,playing-30%,eating-15%,running-5%) and give answer, here add 90% the four value will come
```

```python
# we can use anyone either top_k or top_p.
```

**Prompting Llama 2**

```python
# Prompt Template

# '''
# <s>[INST] <<SYS>>
# {{ system_prompt }}
# <</SYS>>

# {{ user_message }} [/INST] {{model_answer}} </s>
# '''
```

```python
prompt = '''

<s>[INST] <<SYS>>
You are a helpful, respectful and honest assistant. Always answer as helpfully as possible, while being safe.
Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content.
Please ensure that your responses are socially unbiased and positive in nature.
If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct.
If you don't know the answer to a question, please don't share false information.
<</SYS>>
I liked friends and money heist. Recommend me the similar shows to watch.
[/INST]
'''

print(get_response(prompt))
```
